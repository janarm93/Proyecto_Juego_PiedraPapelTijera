```python
import random    #para seleccionar opciones aleatorias para la maquina entre piedra, papel o tijera
import emoji      #para agregar emojis en los prints

# Variables previas
registro_jugadores = {}  #diccionario para almacenar los jugadores y sus estadisticas (partidas ganadas, perdidas o empatadas)
opciones_respuesta = ["piedra", "papel", "tijera"]    #lista que tiene las unicas opciones del juego: piedra, papel o tijera

# Funcion para registrar a un participante, si el jugador ya esta registrado en registro_jugadores muestra un mensaje y si no esta añadido, lo añade con las estadisticas de de cero y le muestra un mensaje.
def vamos_participante(nombre, registro):
    if nombre in registro:
        print(f"\n¡Bienvenido/a de nuevo {nombre}! {emoji.emojize(':grinning_face:')} ya has agotado tus partidas")
    else:
        registro[nombre] = {"ganadas": 0, "perdidas": 0, "empatadas": 0}
        print(f"\n¡Hola, {nombre}! Bienvenido/a al juego {emoji.emojize(':desert_island:')}")

def iniciar_juego():
    # Inicio del juego con gestion de errores, da la bienvenida, pide el nombre al jugador y con .strip elimina los espacios 
    try:
        print("¿Preparado para jugar a Piedra, Papel o Tijera!?")
        nombre_usuario = input("Introduce tu nombre: ").strip().capitalize()

        # Registramos al nuevo usuario  y llama a la funcion
        vamos_participante(nombre_usuario, registro_jugadores)

        # Mensaje inicial
        print(f"""\n {nombre_usuario} \U0001F600
    Hasta ahora eras tú quien cumplía los sueños de tus viajeros. 
    Ahora nos toca a nosotros cumplir el viaje de tus sueños.""")
        print("Piedra, papel, tijera... ¡y vuela! \U0001F6E9")
        print("---------------------------")

        # Variables del juego
        partidas_jugadas = 0     #este es el contador de partidas
        partidas_ganadas_usuario = 0   #el contador de las partidas del usuario
        partidas_ganadas_maquina = 0    #el contador de las maquina
        partidas_empatadas = 0    #el contador de las partidas empatas
        total_partidas = 3     #el paximo de las partidas pemitidas para cada jugador


        # iteracion del juego
        while partidas_jugadas < total_partidas:    #El bucle se ejecuta mientras partidas_jugadas sea menor que total_partidas
            respuesta_usuario = input("Elige piedra, papel o tijera. Introduce 'salir' para terminar el juego: ").lower().strip()
            if respuesta_usuario == "salir":
                print("¡Gracias por jugar! Saliendo del juego...")
                break
            elif respuesta_usuario not in opciones_respuesta:   #Si el usuario escribe algo que no sea "piedra", "papel" o "tijera", se muestra un mensaje de error y vuelve a pedir una entrada válida.
                print("\nOpción incorrecta. Recuerda que sólo puedes elegir entre piedra, papel o tijera. No vale lagarto ni spock")
                print("---------------------------")
                continue
                #logica del juego pura y dura la maquina selecciona una respuesta aleatoria y despues se va incrementando el numero de partidas jugadas para llegar llevar conteo en las estadisticas de cada jugador
            
            respuesta_maquina = random.choice(opciones_respuesta)
            partidas_jugadas += 1
            print(f"\nEsta es la ronda: {partidas_jugadas}")
            print(f"Tu has elegido: {respuesta_usuario}. Tu oponente ha escogido: {respuesta_maquina}.")

            if respuesta_usuario == respuesta_maquina:
                partidas_empatadas += 1
                print(f"\n¡Empate! Lleváis {partidas_empatadas} rondas empatadas.")
                print("---------------------------")
            elif (respuesta_usuario == "piedra" and respuesta_maquina == "tijera") or \
                (respuesta_usuario == "papel" and respuesta_maquina == "piedra") or \
                (respuesta_usuario == "tijera" and respuesta_maquina == "papel"):
                partidas_ganadas_usuario += 1
                print(f"\n¡Has ganado! Llevas {partidas_ganadas_usuario} rondas ganadas.")
                print("---------------------------")
            else:
                partidas_ganadas_maquina += 1
                print(f"\nHas perdido. Tu oponente ha ganado {partidas_ganadas_maquina} rondas.")
                print("---------------------------")

        # Actualizar el registro del jugador con sus estadisticas
        registro_jugadores[nombre_usuario]["ganadas"] += partidas_ganadas_usuario
        registro_jugadores[nombre_usuario]["perdidas"] += partidas_ganadas_maquina
        registro_jugadores[nombre_usuario]["empatadas"] += partidas_empatadas

        # resultados finales 
        print(f"\n--- RESULTADO: ALEA JACTA EST! ---")
        print(f"Partidas jugadas: {partidas_jugadas}")
        print(f"Ganadas por ti: {partidas_ganadas_usuario}")
        print(f"Ganadas por tu oponente: {partidas_ganadas_maquina}")
        print(f"Empates: {partidas_empatadas}")

        # con esto podemos ver quien ha sido el vencedor de la partida tras las 3 rondas
        if partidas_ganadas_usuario > partidas_ganadas_maquina:
            print(f"\n¡Felicidades! Has ganado el juego y puedes participar en el sorteo {emoji.emojize(':desert_island:')}")
        elif partidas_ganadas_usuario < partidas_ganadas_maquina:
            print(f"\nLo siento, {nombre_usuario} La máquina ha ganado el juego {emoji.emojize(':robot:')}")
        else:
            print("\nEl juego ha terminado en empate.")

        # Mostrar estadisticas del usuario
        print("\n--- ESTADÍSTICAS ACTUALES ---")
        print(f"Estadísticas de {nombre_usuario}:")
        print(f"Ganadas: {registro_jugadores[nombre_usuario]['ganadas']}")
        print(f"Perdidas: {registro_jugadores[nombre_usuario]['perdidas']}")
        print(f"Empatadas: {registro_jugadores[nombre_usuario]['empatadas']}")
        return {"nombre": nombre_usuario, "ganadas": registro_jugadores[nombre_usuario]["ganadas"], "perdidas": registro_jugadores[nombre_usuario]["perdidas"], "empatadas": registro_jugadores[nombre_usuario]["empatadas"]}
    except Exception as e:
        print("Algo inesperado ha ocurrido.")
        print(f"Veamos que ha pasado: {e}")
        return None

resultado_final = iniciar_juego()
if resultado_final:
    print("--- Resultados guardados \U0001F504 ---")
```
